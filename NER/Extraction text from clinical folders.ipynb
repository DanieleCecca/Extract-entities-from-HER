## Discharge  letter

To locate the discharge letter within the text, we can use two approaches:
The first is to use a start and end string to identify the document’s content.
The second is to rely solely on the file title, which should be present in the header of each page.

For now, I’m using the second approach; the initial string/title used to identify the file was provided by the doctor.

**Initial string/ID**: *RELAZIONE CLINICA alla DIMISSIONE - Definitiva*

```python
def estrai_relazione_dimissione(file_md_path):
  with open(file_md_path, 'r', encoding='utf-8') as f:
      contenuto = f.read()

  # La regex con lookahead per mantenere il separatore all'inizio di ogni pagina
  # Divide prima di ogni {numero}------------------------------------------------ tranne all'inizio del testo
  pagine = re.split(r'(?=\{\d+\}[-]{48})', contenuto)

  # Filtra le pagine che contengono la stringa specifica
  pagine_filtrate = [pagina for pagina in pagine if 'RELAZIONE CLINICA alla DIMISSIONE' in pagina]

  # Unisci le pagine filtrate
  risultato = '\n\n'.join(pagine_filtrate)

  return risultato
```

```python
def save_lettera_dimissione(lettera_dimissione, output_path,file_md):
  output_path = os.path.join(output_path,f'dimissione_{file_md}')
  with open(output_path, 'w', encoding='utf-8') as f:
    f.write(lettera_dimissione)
```

```python
def estrai_lettera_dimissione_folder(data_path, output_path):
  file_list=os.listdir(data_path)

  for file in file_list:
    folder_path=os.path.join(data_path, file)
    if os.path.isdir(folder_path):
      files=os.listdir(folder_path)
      file_md=[file for file in files if '.md' in file][0]
      file_md_path=os.path.join(folder_path, file_md)

      lettera_dimissione=estrai_relazione_dimissione(file_md_path)
      save_lettera_dimissione(lettera_dimissione,output_path, file_md)
      print(f'File {file_md} processato')
```

```python
estrai_lettera_dimissione_folder(DATA_PATH,OUTPUT_DIMISSIONI)
```
